```python
import albumentations as A
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
transform = A.GaussNoise(std_range=(0.20, 0.22), mean_range = (0.0,0.0),p=0.5)
im = cv2.imread("foto_original.jpg")

image = transform(image=im)['image']

plt.axis("off")
plt.imshow(image)
plt.show()

```
