```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# Imagens originais
Im_1 = Image.open('./Images/Fig0308(a)(fractured_spine) (1).tif')
Im_2 = Image.open('./Images/Fig0309(a)(washed_out_aerial_image).tif')
```

```python
""" 
Para reproduzir as imagens do livro solicitadas na atividade, precisamos aplicar a transformação de potência.
A fórmula utilizada é: s = c * r^γ, em que s é o valor do pixel transformado, r é o valor do pixel original,
γ (gamma) é o expoente da transformação de potência, e c é uma constante de escala.
"""
def power_transform(image, gamma, c):
    # Aplica a transformação de potência
    transformed_image = c * np.power(image, gamma)
    return transformed_image
```

```python
""" 
Para reproduzir as imagens referentes a figura 3.8 do livro, 
precisamos aplicar a transformação de potência descrita anteriormente com os valores de gamma definidos no livro.
"""

images_1 = []
orig_arr = np.asarray(Im_1)
images_1.append(orig_arr)

c = 1.0

for gamma in [0.6, 0.4, 0.3]:
    images_1.append(power_transform(orig_arr, gamma, c))

titles = ['Imagem Original', 'γ = 0.6', 'γ = 0.4', 'γ = 0.3']
plt.figure(figsize=(8, 10))
for i in range(len(images_1)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images_1[i], cmap = 'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
"""
Para reproduzir as imagens referentes a figura 3.9 do livro, 
aplicamos a mesma transformação de potência na segunda imagem, também com os valores de gamma definidos no livro.
"""
images_2 = []
orig_arr_2 = np.asarray(Im_2)
images_2.append(orig_arr_2)

c = 1.0

for gamma in [3.0, 4.0, 5.0]:
    images_2.append(power_transform(orig_arr_2, gamma, c))

titles_2 = ['Imagem Original', 'γ = 3.0', 'γ = 4.0', 'γ = 5.0']
plt.figure(figsize=(8, 10))
for i in range(len(images_2)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images_2[i], cmap = 'gray')
    plt.title(titles_2[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
