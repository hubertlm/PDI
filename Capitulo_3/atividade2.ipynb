```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reproduzindo a imagem 3.34 do livro

Im = cv2.imread('./Images/Fig0334(a)(hubble-original).tif', cv2.IMREAD_GRAYSCALE)

# Para reproduzir a imagem do livro, precisamos implementar um filtro espacial de media 15x15
def spatial_mean_filter(image, filter_size):
    processed_image = cv2.blur(image, (filter_size, filter_size))
    return processed_image

Blur_Im = spatial_mean_filter(Im, 15)

# Visando realizar o ultimo item, precisamos idenficar a intensidade da imagem borrada, 
# Para aplicarmos uma limiarizao para eliminar objetos com base em sua maior intensidade

# Vamos definir um limiar de 25% da intensidade mxima da imagem borrada, seguindo o que foi descrito no livro
threshold = 0.25 * np.max(Blur_Im)
threshold_im = np.where(Blur_Im >= threshold, 255, 0).astype(np.uint8)

plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.title('Imagem Original')
plt.imshow(Im, cmap = 'gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Imagem Processada com Filtro de Mdia')
plt.imshow(Blur_Im, cmap = 'gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Imagem Limiarizada')
plt.imshow(threshold_im, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
# Reproduzindo a imagem 3.35 do livro

Im2 = cv2.imread('./Images/Fig0335(a)(ckt_board_saltpep_prob_pt05).tif', cv2.IMREAD_GRAYSCALE)

# Podemos reutilizar a funo de filtro espacial de media, mas agora com dimenses 3x3
Blur_Im2 = spatial_mean_filter(Im2, 3)

# Agora precisamos estabelecer uma funo para aplicar um filtro de mediana
def spatial_median_filter(image, filter_size):
    processed_image = cv2.medianBlur(image, filter_size)
    return processed_image

Median_Im2 = spatial_median_filter(Im2, 3)

plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.title('Imagem Original')
plt.imshow(Im2, cmap = 'gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Imagem Processada com Filtro de Mdia')
plt.imshow(Blur_Im2, cmap = 'gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Imagem Processada com Filtro de Mediana')
plt.imshow(Median_Im2, cmap = 'gray')
plt.axis('off')
plt.show()
```
