```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Imagem original
Im = cv2.imread('Images/Fig0819(a).tif')


# Função para reduzir a imagem
def reduce(Image, factor):
    factor = factor/
    height, width = Image.shape[:2]
    new_height = height//factor
    new_width = width//factor
    return cv2.resize(Image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)

# Função para amplificar a imagem por replicação de pixels
def ampifly(Image, factor):
    height, width = Image.shape[:2]
    new_height = height*factor
    new_width = width*factor
    return cv2.resize(Image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)

Im_reduced = reduce(Im,10)
Im_back_to_normal = ampifly(Im_reduced,10)

plt.figure(figsize=(15, 5))

# --- Imagem 1: Original ---
plt.subplot(1, 3, 1)     
plt.title('Original')     
plt.imshow(Im, cmap='gray')
plt.axis('off')           

# --- Imagem 2: Reduzida ---
plt.subplot(1, 3, 2)     
plt.title('Reduzida')
plt.imshow(Im_reduced, cmap='gray')
plt.axis('off')

# --- Imagem 3: Ampliada de Volta ---
plt.subplot(1, 3, 3)      
plt.title('Ampliada de Volta')
plt.imshow(Im_back_to_normal, cmap='gray')
plt.axis('off')

plt.show()
```
